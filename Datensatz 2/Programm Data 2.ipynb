Einführung des Programms mit Einlesung der Datei(en)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = pd.read_csv('data-2.csv', sep=';',header=1,decimal=',',skip_blank_lines=True,skipfooter=3, engine='python')  #[seperator muss auf semicolon gestezt werden, da es in den Daten so gemacht wird // ALTE DATEN]

df = pd.DataFrame(data)

data_columns = ['Jahr', 'Export­quote']
num_df = (df.drop(data_columns, axis=1).join(df[data_columns].apply(pd.to_numeric, errors='coerce')))
print(num_df)
  
print(df)
df = num_df
df = df.dropna()
print(df)
```

Speicherung der Variablen

```python
#Speicherung der Variablen
data.to_csv('Urliste_1.csv', index=False, sep=',', decimal='.') # Um im Deutschen System Gelesen zu werden muss die Datei mit ";" als seperator und "," als Decimalpunkt gespeichert werden 
#Speicherung als rangliste
df_Sort = df.sort_values(by=['Export­quote', 'Jahr'], ascending=[True, True])
df_Sort.to_csv('Rangliste_1_ENGLISCH.csv', sep=',', decimal='.')
df_Sort.to_csv('Rangliste_1_DEUTSCH.csv', sep=';', decimal=',')
df.to_excel(excel_writer='Urlise_EXCEL.xlsx')

df_Sort = df.sort_values(by=['Jahr','Export­quote' ], ascending=[True, True])
df_Sort.to_csv('Rangliste_1_Nach_JAHR_ENGLSICH.csv', sep=',', decimal='.')

# WIPPPPPPPP
```

Verschiedene Einzelteile des Datensatzes


```python


print("Quartile und mehr")
print(df.describe())
print("")


print("\nModus")
print(df.mode(axis=0))
print("")

print("Artihmetischer Mittelwert")
print(df.mean())
print("")

print("Median")
print(df.median())
print("\n\n")


print("spannweite")
print(df.max()-df.min())
print("")

print("\nMittler abweichung von median")
print(df.mad())
print("")


#VARIANZ (STICHPROBENVARIANZ)
print("VARIANZ")
print(df.var())
print("")

print("\n\n VARIATIONSKOEFFIZIENT")
cv = lambda x: np.std(x, ddof=1) / np.mean(x) * 100 
print(df.apply(cv))
print("")

print("Covarianz")
print(df.cov())
print("")

print("Correlationskoeffizent")
print(df.corr())
print("")

print("Korrelationskoeffizient nach Spearman")
print(df.corr(method='spearman'))
print("")

#Decile werden ausgegeben 
for x in range(1, 11):
    x = x*0.1
    print("quartil ", x)
    print(df.quantile(x))
    
    
```
